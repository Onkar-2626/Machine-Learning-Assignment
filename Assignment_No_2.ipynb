```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
print('Libraries imported.')
```

```python

url = '/content/sample_data/mnist_test.csv'
columns = ['sepal_length','sepal_width','petal_length','petal_width','class']
df = pd.read_csv(url, names=columns)
df.head()
```

```python

df.info()


```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# Example strategy: fill numeric columns with mean
for col in df.select_dtypes(include=['float64','int64']).columns:
    df[col].fillna(df[col].mean(), inplace=True)

# Fill categorical values with mode
for col in df.select_dtypes(include=['object']).columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

df.isnull().sum()
```

```python
before = df.shape[0]
df.drop_duplicates(inplace=True)
after = df.shape[0]
print('Duplicates removed:', before - after)
```

```python
encoder = LabelEncoder()
df['class'] = encoder.fit_transform(df['class'])
df.head()
```

```python
scaler = StandardScaler()
cols = ['sepal_length','sepal_width','petal_length','petal_width']
df[cols] = scaler.fit_transform(df[cols])
df.head()
```

```python
df.to_csv('assignment2_cleaned_iris.csv', index=False)
print('Saved: assignment2_cleaned_iris.csv')
```

This code performs complete data preprocessing to make the dataset suitable for machine learning. It loads the data, explores its structure, and handles missing values using mean and mode strategies. Duplicate records are removed to improve data quality. Categorical class labels are converted into numerical form using label encoding. Numerical features are standardized to ensure uniform scaling. Finally, the cleaned and processed dataset is saved for further model training and analysis.
